```python
# 1. Limpieza de dependencias conflictivas
!pip uninstall -y ldm
!pip install taming-transformers-rom1504 git+https://github.com/openai/CLIP.git k-diffusion

import os
import re
import numpy
%cd /content/stable-diffusion
os.environ['PYTHONPATH'] = "/content/stable-diffusion"
os.environ['MPLBACKEND'] = "Agg"

def robust_patch():
    f_path = 'frontend/frontend.py'
    if os.path.exists(f_path):
        with open(f_path, 'r') as f: content = f.read()

        # Reemplazos de Gradio 6
        content = content.replace('_js=', 'js=')
        content = content.replace('source="upload"', 'sources=["upload"]')
        content = content.replace('tool="sketch"', '')
        content = content.replace('tool="select"', '')
        content = re.sub(r'\.style\s*\(.*?\)', '', content, flags=re.DOTALL)

        # REPARACIÓN QUIRÚRGICA: Forzamos la lista de outputs de txt2img a ser válida
        # Buscamos el bloque que contiene la lista de outputs problemática y lo reemplazamos por uno limpio
        content = re.sub(
            r'outputs\s*=\s*\[\s*output_txt2img_gallery,\s*output_txt2img_seed,\s*output_txt2img_params,\s*output_txt2img_stats.*?\]',
            'outputs=[output_txt2img_gallery, output_txt2img_seed, output_txt2img_params, output_txt2img_stats]',
            content, flags=re.DOTALL
        )

        # Eliminamos cualquier mención al editor que pueda causar errores de nombre
        content = content.replace('img2img_image_editor.edit()', 'None')

        # Limpieza final de sintaxis
        content = re.sub(r',\s*,', ',', content)
        content = re.sub(r'\[\s*,', '[', content)
        content = re.sub(r',\s*\]', ']', content)

        with open(f_path, 'w') as f: f.write(content)

    u_path = 'frontend/ui_functions.py'
    if os.path.exists(u_path):
        with open(u_path, 'r') as f: content = f.read()
        content = content.replace('_js=', 'js=')
        with open(u_path, 'w') as f: f.write(content)

robust_patch()

# 3. Lanzar la WebUI con parches de seguridad para PyTorch 2.6
ckpt_path = f"{models_path}/sd-v1-4.ckpt"
out_path = f"{output_path}"
launch_args = ["--ckpt", ckpt_path, "--outdir", out_path, "--share"]
launch_args += vars.split()
args_str = str(launch_args)

!python -c "import sys; import torch; import numpy; import pytorch_lightning; import gradio as gr; from pytorch_lightning.utilities import rank_zero; sys.modules['pytorch_lightning.utilities.distributed'] = rank_zero; setattr(gr.components.Component, 'style', lambda x, *args, **kwargs: x); setattr(gr.Row, 'style', lambda x, *args, **kwargs: x); from pytorch_lightning.callbacks.model_checkpoint import ModelCheckpoint; safe_list = [ModelCheckpoint, numpy.core.multiarray.scalar, numpy.dtype, numpy.dtypes.Float64DType, torch.HalfStorage, torch._utils._rebuild_tensor_v2]; torch.serialization.add_safe_globals(safe_list); sys.argv=['scripts/webui.py'] + {args_str}; exec(open('scripts/webui.py').read())"
```
